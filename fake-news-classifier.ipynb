```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, HashingVectorizer
```

```python
df = df.dropna()
```

```python
message = df.copy()
```

```python
message.reset_index(inplace=True)
```

```python
message
```

```python
message['title'][6]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
import re
```

```python
corpus = []
for i in range(len(message)):
    clean = re.sub('[^a-zA-Z]',' ',message['title'][i])
    clean = clean.lower()
    clean = clean.split()
    
    clean = [ps.stem(word) for word in clean if not word in stopwords.words('english')]
    clean = ' '.join(clean)
    corpus.append(clean)
```

```python
cv = CountVectorizer(ngram_range=(1,3), max_features=5000)
x = cv.fit_transform(corpus).toarray()
```

```python
z = message['label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,z,test_size=0.33,random_state=0)
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn import metrics
pred = model.predict(X_test)
score = metrics.accuracy_score(y_test, pred)
```

```python
print("Accuracy: %0.3f" %score)
```

```python
cm = metrics.confusion_matrix(y_test, pred)
```

```python
plt.figure(figsize=(10,10))
import seaborn as sn
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel("Actual")
plt.ylabel("Predicted")
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier
cl = PassiveAggressiveClassifier()
```

```python
cl.fit(X_train,y_train)
```

```python
pred = cl.predict(X_test)
score = metrics.accuracy_score(y_test, pred)
print("Accuracy: %0.3f" %score)
cm = metrics.confusion_matrix(y_test, pred)
plt.figure(figsize=(10,10))
import seaborn as sn
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel("Actual")
plt.ylabel("Predicted")
```
